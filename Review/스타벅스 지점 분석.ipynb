2. 읽어오고 싶은 파일 선택

- 헤더 中 요청 URL

```python
post_url = "https://www.starbucks.co.kr/store/getStore.do?r=QPXZ0WERYL"
```

- 페이로드 中 양식 데이터

```python
payload = {
"in_biz_cds": 0,
"in_scodes": 0,
"ins_lat": 37.473889,
"ins_lng": 127.0292881,
"search_text": '',
"p_sido_cd": '01',
"p_gugun_cd": '',
"in_distance": 0,
"in_biz_cd": '',
"isError": 'true',
"searchType": 'C',
"set_date": '' ,
"all_store": 0,
"T03": 0,
"T01": 0,
"T27": 0,
"T12": 0,
"T09": 0,
"T30": 0,
"T05": 0,
"T22": 0,
"T21": 0,
"T10": 0,
"T36": 0,
"T43": 0,
"T48": 0,
"P10": 0,
"P50": 0,
"P20": 0,
"P60": 0,
"P30": 0,
"P70": 0,
"P40": 0,
"P80": 0,
"whcroad_yn": 0,
"P90": 0,
"new_bool": 0,
"iend": 1000,
"rndCod": 'BYGQQYEIAM',
}
```

```python
r = requests.post(post_url, data=payload)
r.text # -> 문자열 데이터 타입
```

```python
# 딕셔너리 타입으로 전환
r.json()['list']
```

```python
df = pd.DataFrame(r.json()['list'])
df.head(1)
```

2. 스타벅스 이미지 저장하기 (starbucks 폴더 만들고 이미지 저장)

```python
from urllib import request
import os
import random
import time 
```

```python
host = "https://www.starbucks.co.kr"
os.getcwd() # 현재 위치의 directory 보기
# 파이썬에서는 역슬래시를 반드시 두번 해주어야 역슬래시 하나로 인식하다.
```

```python
for idx, row in seoul_star[['s_name', 'defaultimage']].iterrows():
    if os.path.isdir("./starbucks") == False: # isdir : 현재 directory에 있는지
        os.mkdir("./starbucks") # mkdir : 폴더 생성
    request.urlretrieve(host + row['defaultimage'], "./starbucks/{}.jpg".format(row['s_name']) )
```
